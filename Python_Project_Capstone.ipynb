## Task 2:
##### The cost column in the dataframe “Project” has some missing values. Your task is to compute these missing values. Replace the missing values by running average. You should use the “For” loop for this task. 

```python
#Importing the project_dataframe.csv file as DataFrame
#Reading the project dataframe .csv file
project = pd.read_csv("project_dataframe.csv")

print("Project DataFrame: ")
project
```

```python
#Creating a list of running averages using 'Cost' column in the project dataframe

running = 0
count = 0

# Calculating the average cost
average_cost = round(proj_df['Cost'].mean(), 2)

# Replacing missing values with average cost using a for loop
for i in range(len(proj_df)):
    if pd.isnull(proj_df.loc[i, 'Cost']):       #isnull() function is used to check if there is any null values
        proj_df.loc[i, 'Cost'] = average_cost

    #else function is used to iterate the loop if there is no missing values
    else:
        running += proj_df.loc[i, 'Cost']
        count += 1

print("Updated Project DataFrame:")
print(proj_df)
```

## Task 5:

##### Add a new bonus column in the Final dataframe. Give a 5% bonus concerning project cost only to employees who have finished the projects. 

```python
# Creating a copy of final DataFrame
final_copy = final

print("Copy of Final DataFrame: ")
final_copy
```

```python
#Using 'for' loop to creating a list of Bonus cost
#Later, inserted the list into the final dataframe
results_list=[]

for i, r in final_copy.iterrows():
    if(final_copy['Status'].iat[i] == 'Finished'):             #checking every index if "Finished" is present at any index of status column
        result = round((r['Cost']) + (r['Cost']) * 0.05, 2)    #if found, then bonus = cost + 5% of cost also rounding the bonus upto 2 decimal points
        results_list.append(result)                            #append() will add the result in the end of the list
    else:
        result = round((r['Cost']) + (r['Cost']) * 0, 2)       #if not, then no bonus
        results_list.append(result)

final_copy.insert(9, "Bonus-5%", results_list)                 #inserting the bonus-5% column into the final Dataframe at index 9

print("Updated Final DataFrame: ")
final_copy
```

## Task 6:

##### Demote the designation level by 1, whose projects have status “fail”. Delete the employees record whose designation level is above 4. 

```python
#As per the task given, I used 'for' loop to create a list of new designations
#Later,inserted the list into the final dataframe

new_designation = []

for i, r in final_copy.iterrows():
    #checking if the status of the project is 'Failed'...
    if (final_copy["Status"].iat[i] == 'Failed'):
        result = (r['Designation Level']) - 1    #if 'Failed' then demote 1 level from designation level
        new_designation.append(result)
    else:
        result = (r['Designation Level']) - 0    #if not 'Failed' then do not demote any level from designation level
        new_designation.append(result)

#inesrting the new designation column...
final_copy.insert(10, 'Designation', new_designation)     #inserting the Designation column into the final Dataframe at index 10

print("Updated Final DataFrame: ")
final_copy
```

```python
#droping the previous designation level
final_copy.drop("Designation Level", axis=1 , inplace=True)
```

```python
#DataFrame with updated and renamed Designation Levels

final_copy.rename(columns={'Designation': 'Designation Level'}, inplace=True)

print("Updated Final DataFrame: ")
final_copy
```

```python
#deleting the employee records having designation level > 4

to_delete = 0

for index, row in final_copy.iterrows():
    if row["Designation Level"] > 4:                      #checking if any of the employees have designation level > 4
        to_delete.append(index)                           #if found, then adding the index at the end of the list `to_delete`
        final_copy.drop(to_delete, inplace=True)          #then, removing the index from the Dataframe

print("Updated Final DataFrame: ")
final_copy
```
